```python
def get_slice_numbers(dicom_datasets, z_slice):
    # Find actual DICOM slice numbers from z_slice
    slice_positions = [ds.SliceLocation for ds in dicom_datasets]
    sorted_positions = sorted(slice_positions)
    return sorted_positions.index(slice_positions[z_slice.start]), sorted_positions.index(slice_positions[z_slice.stop-1])

get_slice_numbers(ds, slice(0, 10))
```
